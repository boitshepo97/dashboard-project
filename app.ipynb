```python
import dash
import dash_html_components as html
import dash_core_components as dcc
import plotly.graph_objs as go
import plotly.express as px
import dash_table
import seaborn as sns
import pandas as pd 
import numpy as np


df = pd.read_csv('Data_ ReliefWeb Crisis Figures Data - latest_figures.csv')
df.head(1)
```

```python
numbers = df[['crisis_name','figure_name','figure_value']]
numbers
```

```python
# data = df.groupby(['People in Need', 'People Targeted for Assistance']) 
# data

total= df['figure_value'].groupby(df['crisis_name']) 
total.first()
```

```python
options = ['People in Need', 'People Targeted for Assistance'] 
  
# selecting rows based on condition 
r = numbers.loc[numbers['figure_name'].isin(options)]

print(r)
```

```python
app = dash.Dash()

app.layout = dash_table.DataTable(
    id='table',
    columns=[{"crisis_name": i, "id": i} for i in df.columns],
    data=r.to_dict('records'),
)
```

```python
def generate_table(dataframe, max_rows=26):
    return html.Table(
        # Header
        [html.Tr([html.Th(col) for col in r.columns]) ] +
        # Body
        [html.Tr([
            html.Td(r.iloc[i][col]) for col in r.columns
        ]) for i in range(min(len(r), max_rows))]
    )
```

```python
app.layout = html.Div([
                   
                    dcc.Graph(id='stacked bargraph',
                        figure = px.bar(r, x="figure_name", y="figure_value", 
                            height=10000, 
                            title='Numbers according to data')),    

                    dcc.Graph(id='grouped bargraph', 
                        figure = px.bar(df, x="crisis_name", y="figure_value", color='figure_name', 
                            barmode ='group', height=650, 
                            title='crisis_ios3')),   

                    dcc.Interval(id='interval-component', 
                        interval=2000,
                        n_intervals=0)




                    ])
```

```python
if __name__ == '__main__':
    app.run_server(debug=True)
```
